```python
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv("C:/MRT/clusters_5/noTime.csv")
weather.head()
```

```python
weather_count=weather.drop(['Unnamed: 0'],axis=1)
weather_dropna=weather_count.dropna()
weather_dropna
```

```python
X=weather_dropna.iloc[:, 1:8].values
print(X)
# X=X.to_numpy()
y=weather_dropna.target.values
# y=y.to_numpy()
```

```python
pca = PCA(n_components=2) # n_components 為要降到的維度，這邊降為二維
pca.fit(X) # 使用 x 為來源資料
result = pca.transform(X) # 使用 transform() 即可取得降維後的陣列
print(result)
```

```python
plt.scatter(result[:,0], result[:,1], c=y, s=25, alpha=0.4, marker='o')
```
